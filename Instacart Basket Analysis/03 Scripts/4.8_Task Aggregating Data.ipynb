```python
# creating a column to record maximum number of orders by each customer
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform('max')
```

```python
# Deriving a flag column based on max_order column
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

#### 06.

```python
# deriving a variable based on customer's order prices
ords_prods_merge['Average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
ords_prods_merge.head(60)
```

```python
# creating a flag column based on average price
ords_prods_merge.loc[ords_prods_merge['Average_price'] < 10, 'spending_flag'] = 'Low spender'
ords_prods_merge.loc[ords_prods_merge['Average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.drop_duplicates('user_id')['spending_flag'].value_counts()
```

```python
total_users = ords_prods_merge['user_id'].nunique()
```

```python
total_users
```

```python
# performing sainty check using quantiles
ords_prods_merge['Average_price'].describe()
```

##### looking at the statistics above we can see that our threshold of 10 is too high thats why a huge number of customers are labeled as low spender.

```python
ords_prods_merge = ords_prods_merge.drop('spending_flag', axis = 1)
```

```python
# creating spending flag using 75th percentile
threshold = ords_prods_merge['Average_price'].quantile(0.75)  #  8.23
```

```python
ords_prods_merge['spending_flag'] = ords_prods_merge['Average_price'].apply(
    lambda x: 'High spender' if x >= threshold else 'Low spender'
)
```

```python
ords_prods_merge.drop_duplicates('user_id')['spending_flag'].value_counts()
```

```python
ords_prods_merge[['user_id', 'spending_flag', 'Average_price']].head(60)
```

#### 07.

```python
# deriving a variable based on customer's order frequency
ords_prods_merge['median_days_between_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
ords_prods_merge.head(60)
```

```python
# creating a flag column based on order frequency
# Flag for non-frequent customers
ords_prods_merge.loc[ords_prods_merge['median_days_between_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'

# Flag for regular customers
ords_prods_merge.loc[(ords_prods_merge['median_days_between_orders'] > 10) & (ords_prods_merge['median_days_between_orders'] <= 20), 'order_frequency_flag'] = 'Regular customer'

# Flag for frequent customers
ords_prods_merge.loc[ords_prods_merge['median_days_between_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[ords_prods_merge['order_frequency_flag'].isna()]
```

##### We have 5 rows which didnot get any order_frequency_flag, upon inspecting we find that there is one customer who has only placed one order hence NaN in days_since_prior order and thts why no median_days_since_prior_order are calculated for this customer.

```python
# Handle customers with only one order (no median available)
ords_prods_merge.loc[ords_prods_merge['median_days_between_orders'].isna(), 'order_frequency_flag'] = 'One-time customer'
```

```python
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'order_frequency_flag', 'median_days_between_orders']].head(60)
```
